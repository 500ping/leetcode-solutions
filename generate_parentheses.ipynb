```python
 def generateParenthesis(n: int):
    result = []
    
    if n == 0:
        return result
    
    def isValid(s: str) -> bool:
        result = False

        if len(s) <= 1:
            return result

        for i in range(int(len(s) / 2)):
            s = s.replace('()', '')

        if not s:
            result = True

        return result
    
    result = [['(', ')']] * n * 2
    import numpy
    result = [''.join(list(x)) for x in numpy.array(numpy.meshgrid(*result)).T.reshape(-1,len(result)) if isValid(''.join(list(x)))]
    
    return result
```

```python
generateParenthesis(8)
```
